```python
# Step 1: Import Required Libraries
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
from wordcloud import WordCloud

```

```python
# Download NLTK data
nltk.download('stopwords')
nltk.download('punkt')
```

```python
# Step 2: Load the Datasets
# Adjust the file paths if they are different
train_df = pd.read_csv('/content/twitter_training.csv')
valid_df = pd.read_csv('/content/twitter_validation.csv')
```

```python
# Display the first few rows to verify column names
print("Training Dataset Columns:", train_df.columns)
print("Validation Dataset Columns:", valid_df.columns)
```

```python
# Step 3: Data Preprocessing
stop_words = set(stopwords.words('english'))
stop_words
```

```python
def clean_text(text):
    """Function to clean tweet text by removing URLs, special characters, and stopwords."""
    text = re.sub(r'http\S+', '', text)  # Remove URLs
    text = re.sub(r'[^A-Za-z\s]', '', text)  # Remove special characters
    text = text.lower()  # Convert to lowercase
    text = ' '.join([word for word in text.split() if word not in stop_words])  # Remove stopwords
    return text
```

```python
# Convert the last column to strings and handle missing values
train_df['Cleaned_Text'] = train_df.iloc[:, -1].astype(str).apply(clean_text)  # Clean the tweet text

train_df['Cleaned_Text']
```

```python
valid_df['Cleaned_Text'] = valid_df.iloc[:, -1].astype(str).apply(clean_text)  # Clean the tweet text
valid_df['Cleaned_Text']
```

```python
# Step 4: Convert Sentiments to Numerical Labels
# Mapping sentiments based on the third column which seems to contain the sentiment labels
sentiment_mapping = {'Positive': 1, 'Negative': 0, 'Neutral': 2, 'Irrelevant': 3}
train_df['Sentiment_Label'] = train_df.iloc[:, 2].map(sentiment_mapping)
train_df['Sentiment_Label']
```

```python
valid_df['Sentiment_Label'] = valid_df.iloc[:, 2].map(sentiment_mapping)
valid_df['Sentiment_Label']
```

```python
# Check if there are any unmapped sentiment labels
print("\nUnique sentiment labels in training set:", train_df.iloc[:, 2].unique())
print("Unique sentiment labels in validation set:", valid_df.iloc[:, 2].unique())
```

```python
# Step 5: Visualize Sentiment Distribution
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.countplot(x=train_df['Sentiment_Label'].map({1: 'Positive', 0: 'Negative', 2: 'Neutral', 3: 'Irrelevant'}))
plt.title('Training Set Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')

```

```python
plt.subplot(1, 2, 2)
sns.countplot(x=valid_df['Sentiment_Label'].map({1: 'Positive', 0: 'Negative', 2: 'Neutral', 3: 'Irrelevant'}))
plt.title('Validation Set Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Step 6: Feature Extraction using TF-IDF
vectorizer = TfidfVectorizer(max_features=5000)
X_train = vectorizer.fit_transform(train_df['Cleaned_Text'])
y_train = train_df['Sentiment_Label']
X_valid = vectorizer.transform(valid_df['Cleaned_Text'])
y_valid = valid_df['Sentiment_Label']
```

```python

# Step 7: Train the Model
model = MultinomialNB()
model.fit(X_train, y_train)
```

```python

# Step 8: Validate the Model
y_pred = model.predict(X_valid)
```

```python
y_pred
```

```python
# Step 9: Evaluate the Model
print("\nAccuracy on Validation Set:", accuracy_score(y_valid, y_pred))
print("\nClassification Report on Validation Set:\n", classification_report(y_valid, y_pred))

```

```python
# Step 10: Confusion Matrix
cm = confusion_matrix(y_valid, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Negative', 'Positive', 'Neutral', 'Irrelevant'],
            yticklabels=['Negative', 'Positive', 'Neutral', 'Irrelevant'])
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```
